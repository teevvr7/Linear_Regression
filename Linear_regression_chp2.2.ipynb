## 4 : Data Preparation

```python
cars_data
```

```python
X = cars_data[['HP','VOL','SP','WT']]
X
```

```python
y = cars_data['MPG']
y
```

#### There is nothing to prepare because all feature are numerical, but in future we will apply the feature engineering

## 5. Model Building

```python
linear_model = LinearRegression()

# to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.
# this alogrithm will take the best fitted line
```

## 6. Model Training

```python
linear_model.fit(X,y)

# X,y from the data preparation
# X - input
# y - output
```

```python
y # actual output value
```

```python
# Deliverables
```

```python
linear_model.intercept_

# 4 input intercept meet at same point when x equal zero
```

```python
linear_model.coef_

# 4 slope values of the inputs
```

## 7. Model Testing

```python
y
```

```python
y_pred = linear_model.predict(X)
y_pred # predicted output values
```

## 8. Model Evaluation

```python
# actual output - predict output

error = y - y_pred
error
```

## Test 5 : Homoscadacity Test
